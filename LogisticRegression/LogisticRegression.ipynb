## Part 2: Method Justification

When dealing with binary output, 0 or 1, Yes or No, or outcome with multiple categories (e.g. worst, bad, good, better, best), one predictive model commonly implemented is Logistic Regression. It is possible to use the binary logistics regression to model the customer churn behavior. Provided below is the Logistic Regression function:

<b>P (Y=1|X)=  1 / (1+ e^(-1x(β0+β1X1+β2X2+β3X3+β4X4+⋯+βnXn)) ) </b>

where:<br>
β0 to βn are various coefficients <br>
X0 to Xn are the variables impacting the final outcome<br>
And P (Y =1 | X) is the probability of a positive outcome.<br>

Notice the exponent in the function. This is where linear regression plays in (β0+β1X1+β2X2+β3X3+β4X4+⋯+βnXn).

In case of customer churn analysis, for example, one of the X variables can be the “annual income” or the “gender” of the customer. The outcome of the function will tell us how income and or gender determine the probability of service discontinuation by the customer.

Logistic regression function is a sigmoid function (graph below). As can be seen in the graph, the function has output values between 0 and 1. This characterstic of the function helps with prediction of binary outcomes. Logistic regression is binary classification algorithm belonging to generalized linear regression model. It can also be used to solve problem with more than 2 classes.


Let us generate a sigmoid graph below and see how it can be used to predict a binary outcome. To generate the graph, I am using Python code. As can be seen, in the graph, there are 2 levels with a transition from lower to upper level at Y = 0.5. All X-axis values below 0 results in Y values less than 0.5, and all values above 0 form the curve above Y = 0.5. The region of X values below 0 is where negative outcome is assumed or predicted ( 0 or No or negative outcome) and region of X values above 0 is where positive binary outcome is predicted ( 1 or Yes or positive outcome ). Thus, using sigmoid function with linear regression equation in the exponent, we can predict binary dependent variable.

```python
# Code to generate a sigmoid graph showing how the signoid function logic can be used for binary classification

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def sigmoid(x):
    y = []
    for val in x:
        y.append(1/(1+np.exp(-val)))
    return y                  # Y axis values are calculated and using sigmoid function and retured in an array

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10., 10., 0.1)  # Values of X axis between -10 and 10 at intervals of 0.1
y = sigmoid(x)
plt.plot(x,y)                  # Generate graph using matplotlib.pyplot
plt.show()
```

#### A few assumptions are made when using logistic regression. These assumptions are:
1.	Binary logistic regression requires dependent variable to be binary and to follow binomial distribution (e.g. will customer discontinue the service or not, Yes or No). For more than 2 outcomes (ordinal) logistics regression requires the dependent variable categories to be mutually exclusive and exhaustive.
2.	Observations should be independent of each other (e.g. data of one customer should not depend on data of other customer, or same customer should not be used repeatedly in the data)
3.	Multicollinearity among the independent variables should not exists (e.g. avoid using variables from the customer data which depend on each other, say, City, State, County and Zip Code are not all independent)
4.	Linearity of independent variables with respect to log odds of dependent variable (e.g. log odds of the probability of customer discontinuing the service should be linearly related to various variables like gender, income etc)
5.	Large sample size (e.g. customer churn data has 10000 records)

```python
# Following columns have categorical data types and they need to be converted to dummy variables
dummy_variable_columns = ['State',
                          'Churn','Area','Marital','Gender','Techie','Contract','Port_modem','Tablet',
                         'InternetService','Phone', 'Multiple','OnlineSecurity','OnlineBackup',
                          'DeviceProtection','TechSupport', 'StreamingTV','StreamingMovies',
                          'PaperlessBilling','PaymentMethod'
                         ]


# Following columns have continuous numerical data, which needs to be scaled before predictive model can be created.
# Hence, these columns will be transformed to normalized form.

to_transform = ['Population','Children','Age','Income','Outage_sec_perweek','Email',
                'Contacts','Yearly_equip_failure','Tenure',
                'MonthlyCharge','Bandwidth_GB_Year',
                'Timely_response','Timely_fixes','Timely_replacements',
                'Reliability','Options',
                'Respectful_response','Courteous_exchange',
                'Evidence_of_active_listening'
               ]

dummy_data_file_index = 0
```

```python
from sklearn.preprocessing import StandardScaler

def get_dummy_data_with_output(dummy_variable_columns, data):
    global dummy_data_file_index
    dummy_data = pd.get_dummies(data, prefix=dummy_variable_columns, columns=dummy_variable_columns, drop_first= True)
    dummy_data = column_name_remove_space(dummy_data)
    dummy_data.to_csv('dummy_var_data'+str(dummy_data_file_index)+'.csv', index=False)
    y = dummy_data['Churn_Yes']
    dummy_data.drop(columns=['Churn_Yes'], inplace=True)
    sc = StandardScaler()
    dummy_data.loc[:,to_transform] = sc.fit_transform(dummy_data.loc[:,to_transform])
    dummy_data_file_index += 1
    return y, dummy_data

y, dummy_data = get_dummy_data_with_output(dummy_variable_columns, data)
```

## Part 4: Model Comparison and Analysis

### Initial Model

To check the impact, importance and significance of various data columns w.r.t. churn analysis, initial model containing all variables will be created. We have prepared the data in previous step, in a way that it include all the variables at this time.

Let's start with importing the required librarys. We are using sklearn and statsmodel to create models to check how predictive model works and how accurately are we able to predict the outcomes or the customer churn. As we have sufficient number of rows in the data, set I am choosing 70% of the data to be in the training set and 30% of the data to be in the test set.


```python
# Starting with imports

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import statsmodels.api as sm
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# Split the training and test set 7:3 
X_train, X_test, y_train, y_test = train_test_split(dummy_data,y,test_size=.3, random_state=0)
```

```python
# Creating a function to create prediction model based on statsmodel library and
# print details like model Summary, Confusion Matrix and Accuracy Score based on predicted values using test set

def get_sm_model(y_train,X_train, X_test):
    X_train = sm.add_constant(X_train)
    X_test = sm.add_constant(X_test)
    logit_model=sm.Logit(y_train,X_train)
    model=logit_model.fit()
    print(model.summary())
    y_hat = list(map(round, model.predict(X_test)))
    arr = np.c_[y_hat, y_test]
    print('\nPrinting predicted and actual values:\n',arr)
    print('Confusion Matrix \n',confusion_matrix(y_test, y_hat))
    print('Accuracy Score: ',accuracy_score(y_test, y_hat))
    return model, arr

# Creating a function to create prediction model based on sklearn library and
# print details like model Summary, Confusion Matrix and Accuracy Score based on predicted values using test set

def get_sk_model(y_train,X_train, X_test):
    model = LogisticRegression(max_iter=100000000)
    model.fit(X_train, y_train)
    arr = np.c_[X_train.columns.tolist(),model.coef_[0]]
    intercept = model.intercept_[0]
    print('\nPrinting model coefficients and intercept summary for sklearn model:\n',arr, model.intercept_)
    y_pred = model.predict(X_test)
    print('\nPrinting predicted and actual values from sklearn:\n',np.c_[y_pred, y_test])
    print('Confusion Matrix from sklearn\n',confusion_matrix(y_test, y_pred))
    plt.matshow(confusion_matrix(y_test, y_pred))
    plt.title('Confusion matrix')
    plt.colorbar()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.grid(b=None)
    plt.show()
    print('Accuracy Scores from sklearn:\n',accuracy_score(y_test, y_pred))
    print('Classification Report from sklearn:\n',classification_report(y_test, y_pred))
    return model, arr, intercept


# Function to create AUC chart and print AUC score.
# AUC score closer to 1 is considered a good model
# Auc score of 0.5 represents are model which is no
# better than a random guess with 50% chance of predicting the right outcome

from sklearn.metrics import roc_curve
def roc(model, X_test, y_test):
    probs = model.predict_proba(X_test)
    fpr, tpr, _ = roc_curve(y_test, probs[:,1])
    plt.plot(fpr, tpr, marker='.')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()
    print('AUC: %.3f' % roc_auc_score(y_test, probs[:,1]))
```

```python
# Run statsmodel function
get_sm_model(y_train,X_train, X_test);
# Run sklearn model
model, arr, intercept = get_sk_model(y_train,X_train, X_test);
# Run AUC
roc(model, X_test, y_test);
```

### Analysis of statsmodels summary and variable significance

Printed above are the details from predictive models created using statsmodels and sklearn. In the model summary generated by statsmodels, we can see the p-value of various variables and determine which variables are significant for predictive modeling. This will help determine list of variables which can be safely removed without any impact on the overall accuracy and confusion matrix.

### Equation and interpretation of coefficients
Logistic regression equation is:

P (Y=1|X) = exp(y) / 1 + exp(-y)

where y = β0 + β1X1 + β2X2 + β3X3 + ... βnXn

where n is the number of variables in the model and βn is the value of nth coefficient.

In the Logistic regression model, we created above value of n is 26, including all the dummy variables.
<u>For visual sanity, coefficients will be rounded to 2 digital after decimal, to make the the overall equation readable.</u>

<b>Equation with necessary changes to make it readable is:</b>

P (Y=1|X) = 1 / (1 + <b>exp(-1x( -3.15 x Tenure  +0.09 x Contacts  +2.28 x MonthlyCharge  +0.19 x Bandwidth_GB_Year  -0.02 x Age  +0.0 x Children  +1.04 x Techie  -3.24 x Contract_One_year  -3.33 x Contract_Two_Year  +0.05 x Marital_Married  +0.08 x Marital_Never_Married  +0.13 x Marital_Separated  +0.22 x Marital_Widowed  +0.19 x Port_modem  -2.2 x InternetService_Fiber_Optic  -0.62 x InternetService_None  -0.37 x Phone  +0.44 x StreamingMovies  +0.15 x PaperlessBilling  +0.25 x PaymentMethod_Credit_Card_automatic  +0.62 x PaymentMethod_Electronic_Check  +0.22 x PaymentMethod_Mailed_Check  -1.09))</b>)

#### Another way to write this equation is:

Odds ratio = P / ( 1 - P ) = exp(-3.15 x Tenure +0.09 x Contacts +2.28 x MonthlyCharge +0.19 x Bandwidth_GB_Year -0.02 x Age +0.0 x Children +1.04 x Techie -3.24 x Contract_One_year -3.33 x Contract_Two_Year +0.05 x Marital_Married +0.08 x Marital_Never_Married +0.13 x Marital_Separated +0.22 x Marital_Widowed +0.19 x Port_modem -2.2 x InternetService_Fiber_Optic -0.62 x InternetService_None -0.37 x Phone +0.44 x StreamingMovies +0.15 x PaperlessBilling +0.25 x PaymentMethod_Credit_Card_automatic +0.62 x PaymentMethod_Electronic_Check +0.22 x PaymentMethod_Mailed_Check -1.09)

Using this form of the equation we can easily interpret the impact of variables on the odds ratio. Odds ratio for churn is defined a probability of leaving the company divided by probability of not leaving the company. Reducing odds ratio is favorable case for us, as it means probability of discontinuing the service is going down (positive impact on churn, as we want to keep the customer). An increase in odds ratio would be negative impact as customer is more likely to leave the company than to remain with it. We will see how unit change in different variables can impact the odds ratio. 

#### Interpretation of coefficients

Individual coefficients can be interpreted as change in odds ratio for every unit change in the variable. Due to normalization done on the continuous numerical variable, the unit change for such variable is equal to standard deviation. For categorical variable the change is a switch from 0 to 1.

I am calculating the impact of unit increase or decrease (unnormalized form of variable) in the value of independent variable in terms of reduction in odds ratio.

```python
# Function to calculate the reduction percentage in odds ratio
def impact(coeff, field,x, categorical=False):
    if categorical:
        exp = np.exp(x*coeff)
        print ( field, ': Odds Ratio Reduction : '+str(round((1 - exp)*100,2 ))+'%')
    else:
        exp1 = np.exp(x*coeff/data_reduced['Tenure'].std())
        exp2 = np.exp(-1*x*coeff/data_reduced['Tenure'].std())
        print ( field, ': Odds Ratio Reduction with unit increase: '+str(round((1 - exp1)*100,2 ))+'%', 'Odds Ratio Reduction with unit decrease: '+str(round((1 - exp2)*100,2 ))+'%')

        
impact(-3.15,'Tenure',1)
impact(2.28,'MonthlyCharge',1)
impact(-0.02,'Age',1)
impact(-2.2,'InternetService_Fiber_Optic',1,True)
impact(-0.62,'InternetService_None',1,True)
impact(-3.24,'Contract_One_year',1,True)
impact(-3.33,'Contract_Two_Year',1,True)
impact(-0.37,'Phone',1,True)
```

Looking at the coefficients, we can see how a particular variable can impact the churn.  Let's take a few examples, and understand in detail:

1. For variable Tenure , we see a negative coefficient (-3.15 x Tenure). If the user has spent long time with the company (more than mean or average), he is less likely to discontinue, but if he has spent less time with company he is more likely to churn. Mean tenure is 34.53 months. Tenure has standard deviation of 26.44 months. For a customer, with every extra month spent with the company, the odds ratio of leaving the company goes down by 11.2% (1-exp(-3.15/26.44).

2. For variable MonthlyCharge, coeffient is 2.28. With every dollar increase in monthly charge, you make the customer more likely to leave the service. Conversely, with every dollar decrease in monthly charge customer is more likely to stay. With a dollar reduction in monthly charge, odds ratio goes down by 8.26%. With a discount of dollar 5 per month, odds ratio goes down by 35%. With \$10 per month discount the odds ratio goes down by 57.78%.

3. Customers switching from DSL service (default) to fiber optic internet service, will reduce odds ration by 88.92%.

4. Customers switching from month to month contract (default) to one year contract will reduce odd ratio by 96.08%. And those switching to 2 year contract reduce odds ration by 96.42%.
